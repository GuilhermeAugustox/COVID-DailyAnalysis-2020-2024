```python
por_regiao = df['WHO Region'].value_counts()
por_regiao
```

```python
por_regiao.plot(kind='bar')

plt.title('Quantidade de países por região')
plt.xlabel('Regiões')
plt.ylabel('Quantidades de países')

plt.show()
```

```python
total_por_regiao = df.groupby('WHO Region')['Cases - cumulative total'].sum()
total_por_regiao
```

```python
total_por_regiao.plot(kind='bar')

plt.title('Total de Casos por Região')
plt.xlabel('Regiões')
plt.ylabel('Total de Casos')

plt.show()
```
