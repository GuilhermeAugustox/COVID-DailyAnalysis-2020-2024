Separação das colunas à serem utilizadas

```python
df_total = df.drop(columns=['Cases - cumulative total per 100000 population', 'Cases - newly reported in last 7 days', 'Cases - newly reported in last 7 days per 100000 population', 'Cases - newly reported in last 24 hours', 'Deaths - cumulative total per 100000 population', 'Deaths - newly reported in last 7 days', 'Deaths - newly reported in last 7 days per 100000 population', 'Deaths - newly reported in last 24 hours'])

df_total
```

Total de Casos por Região

```python
total_casos_regiao = df_total.groupby('WHO Region')['Cases - cumulative total'].sum()
total_casos_regiao
```

```python
total_casos_regiao_sorted = total_casos_regiao.sort_values(ascending=False)
total_casos_regiao_sorted.plot(kind='bar')

plt.title('Total de Casos por Região')
plt.xlabel('Regiões')
plt.ylabel('Total de Casos')

plt.show()
```

Total de Mortes por Região

```python
total_mortes_regiao = df_total.groupby('WHO Region')['Deaths - cumulative total'].sum()
total_mortes_regiao
```

```python
total_mortes_regiao_sorted = total_mortes_regiao.sort_values(ascending=False)
total_mortes_regiao_sorted.plot(kind='bar')

plt.title('Total de Mortes por Região')
plt.xlabel('Regiões')
plt.ylabel('Total de Casos')

plt.show()
```

Total de Casos por País

```python
total_casos_pais = df_total.groupby('Name')['Cases - cumulative total'].sum()
total_casos_pais
```

```python
total_casos_pais_sorted = total_casos_pais.sort_values(ascending=False)
maiores_10_pais = total_casos_pais_sorted.iloc[1:11]
maiores_10_pais.plot(kind='bar')

plt.title('Os 10 Países com Mais Casos')
plt.xlabel('Países')
plt.ylabel('Total de Casos')
plt.show()
```

Total de Mortes po País

```python
total_mortes_pais = df_total.groupby('Name')['Deaths - cumulative total'].sum()
total_mortes_pais
```

```python
total_mortes_pais_sorted = total_mortes_pais.sort_values(ascending=False)
maiores_10_pais = total_mortes_pais_sorted.iloc[1:11]
maiores_10_pais.plot(kind='bar')

plt.title('Os 10 Países com Mais Mortes')
plt.xlabel('Países')
plt.ylabel('Total de Mortes')
plt.show()
```

Relação entre Casos X Mortes por País

```python
total_casos_mortes_sorted = df.sort_values(by = 'Cases - cumulative total', ascending=False)
top_10 = total_casos_mortes_sorted.iloc[1:11]

casos = top_10['Cases - cumulative total']
mortes = top_10['Deaths - cumulative total']
paises = top_10['Name']

plt.figure(figsize=(10,6))
plt.scatter(casos, mortes, color='b')

for i, pais in enumerate(paises):
    plt.text(casos.iloc[i], mortes.iloc[i], paises.iloc[i], fontsize=9)

plt.title('Relação entre Casos e Mortes por País')
plt.xlabel('Total de Casos')
plt.ylabel('Total de Mortes')
plt.show()
```

```python
total_casos_mortes_sorted = df.sort_values(by = 'Cases - cumulative total', ascending=False)
top_10 = total_casos_mortes_sorted.iloc[1:11]

casos = top_10['Cases - cumulative total']
mortes = top_10['Deaths - cumulative total']
paises = top_10['Name']

bar_with = 0.35
indice = np.arange(len(paises))

plt.figure(figsize = (12, 6))
plt.bar(indice, casos, bar_with, label = 'Casos', color = 'b')
plt.bar(indice + bar_with, mortes, bar_with, label = 'Mortes', color = 'r') 

plt.xlabel('Países')
plt.ylabel('Total')
plt.title('Relação entre Casos e Mortes por País')
plt.xticks(indice + bar_with / 2, paises, rotation = 90)

plt.legend()
plt.show()
```
